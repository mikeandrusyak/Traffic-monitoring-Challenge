```python
import cv2
from tracker2 import*
import time
from picamera2 import Picamera2
```

```python
try:
    if 'picam2' in locals() or 'picam2' in globals():
        picam2.stop()
        picam2.close()
        print("The previous copy of the camera is closed")
except:
    print("There is no previous copy of the camera, or it is already closed.")
```

```python
picam2 = Picamera2()
```

```python
config = picam2.create_still_configuration()
picam2.configure(config)
```

```python
import matplotlib.pyplot as plt

picam2.start()
time.sleep(2)

frame = picam2.capture_array()

frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

plt.imshow(frame_rgb)
plt.axis('off')  
plt.show()

picam2.stop()
picam2.close()
```
